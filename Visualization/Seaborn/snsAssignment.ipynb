**Que 1:** Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**Answer:**

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating visually appealing statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. **Histogram**: A histogram is a plot that displays the distribution of a continuous variable by dividing it into bins and showing the count or frequency of data points within each bin. It is useful for visualizing the overall shape and range of a dataset.

2. **Bar Plot**: A bar plot is a plot that represents categorical data using rectangular bars. It is useful for comparing the values of different categories or groups. Seaborn's bar plot can also display additional statistical information, such as confidence intervals.

3. **Box Plot**: A box plot, or box-and-whisker plot, is a plot that displays the summary statistics of a dataset. It shows the median, quartiles, and any potential outliers. It is useful for visualizing the distribution and identifying outliers in a dataset.

4. **Scatter Plot**: A scatter plot is a plot that represents the relationship between two variables. It displays individual data points as markers on a two-dimensional plane. It is useful for examining the correlation or patterns between two continuous variables.

5. **Heatmap**: A heatmap is a plot that uses colors to represent the values of a two-dimensional dataset. It is particularly useful for visualizing correlation matrices or displaying the intensity of values across categories or time. Seaborn's heatmap also includes features like hierarchical clustering and data normalization.

These are just a few examples of the plots available in Seaborn. The library offers a wide range of other plot types, such as line plots, violin plots, pair plots, and more. The choice of plot depends on the data and the specific insights or relationships you want to visualize.

**Que 2:** Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

**Answer:**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and labels
plt.title("fMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

**Que 3:** Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

**Answer:**


```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data = titanic_data, x='pclass', y='age')
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Age Box Plot")

plt.subplot(1, 2, 1)
sns.boxplot(data = titanic_data, x='pclass', y='fare')
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Fare Box Plot")

plt.tight_layout()
plt.show()
```

**Que 4:** Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

**Answer:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for price with differentiation based on cut
plt.figure(figsize=(8, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title("Histogram of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```

**Que 6:** Use the "flights" dataset from seaborn to plot a heatmap.

**Answer:**


```python
import seaborn as sns

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_matrix = flights_data.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(data=flights_matrix, cmap="YlGnBu")

# Set plot title and labels
plt.title("Passenger Numbers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
