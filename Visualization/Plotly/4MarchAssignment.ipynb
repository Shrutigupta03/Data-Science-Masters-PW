**Q1.** Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

**Answer:**


```python
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot scatter plot using Plotly Express
fig = px.scatter(titanic_data, x="age", y="fare")

# Show the plot
fig.show()
```

**Q2.** Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

**Answer:**

```python
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_data = px.data.tips()

# Plot box plot using Plotly Express
fig = px.box(tips_data, x="day", y="total_bill")

# Show the plot
fig.show()
```

**Q3.** Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

**Answer:**

```python
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_data = px.data.tips()

# Plot histogram using Plotly Express
fig = px.histogram(tips_data, x="sex", y="total_bill", histfunc="count",
                   pattern_shape="smoker", color="day")

# Show the plot
fig.show()
```

**Q4.** Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

**Answer:**


```python
import plotly.express as px

# Load the "tips" dataset from Plotly
iris_data = px.data.iris()

# Plot histogram using Plotly Express
fig = px.scatter_matrix(iris_data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")

# Show the plot
fig.show()
```

**Q5.** What is Distplot? Using Plotly express, plot a distplot.

**Answer:**

In Plotly Express, the distplot function is used to plot a histogram and a kernel density estimation (KDE) plot together, providing a visual representation of the distribution of a dataset. It combines the information from both the histogram and KDE to give a more comprehensive understanding of the data's distribution.

To plot a distplot using Plotly Express, you can follow these steps:

```python
import plotly.express as px

# Load example data
tips_data = px.data.tips()

# Plot distplot using Plotly Express
fig = px.histogram(tips_data, x="total_bill", marginal="rug", nbins=30, histnorm="density")

# Show the plot
fig.show()
```
